```python
import torch
import torch.nn as nn
max_sequence_length = 10
d_model = 6
```

```python
i_even = torch.arange(0, d_model, 2)
i_odd = torch.arange(1, d_model, 2)
denominator_even = torch.pow(10000, (i_even.float() / d_model))
denominator_odd = torch.pow(10000, (i_odd.float() / d_model))


position = torch.arange(max_sequence_length, dtype= torch.float).reshape(max_sequence_length, 1)



```

```python
position
```

```python
even_PE = torch.sin(position/denominator_even)
odd_PE = torch.cos(position/denominator_odd)
```

```python
stacked = torch.stack([even_PE, odd_PE], dim =2)
stacked.shape
```

```python
PE = torch.flatten(stacked,start_dim = 1, end_dim = 2)
PE
```
